```python
#Take two user inputs representing numbers. Detect if they are integers or floats, then print the exact data type of each and their product.
input1 = (input("Enter number: "))
input2 = (input("Enter  number: "))

if '.' in input1:
    input1 = float(input1)
else:
    input1 = int(input1)


if '.' in input2:
    input2 = float(input2)
else:
    input2 = int(input2)
    

print("First input type is ",type(input1))
print("Second input type is ",type(input2))



result = input1 * input2 
print("the result of multiplicated is : " ,result)

```
