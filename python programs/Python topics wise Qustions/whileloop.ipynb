```python
# Q1 Print Fibonacci numbers up to a limit given by the user.
Limitstartnumber = int(input("Enter a number where you decide to start series: "))
Limitendnumber = int(input("Enter a number where you decide to end series: "))
seriesgapnumber = int(input("Enter a number where you decide to gap series: "))
for series in range(Limitstartnumber,Limitendnumber,seriesgapnumber):
    print(series)
```
