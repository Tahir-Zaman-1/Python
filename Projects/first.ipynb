```python
#write a program that  take input marks of three subject and calculate total and average 

#but i alredy took 5 subjects marks there is no problem i will do it with 5 subjects marks

Subject_English = int(input("Entre your English Subjects Marks Out of 100 : "))
Subject_Math = int(input("Entre your Math Subjects Marks Out of 100 : "))
Subject_Science = int(input("Entre your Science Subjects Marks Out of 100 : "))
Subject_Computer = int(input("Entre your Computer Subjects Marks Out of 100 : "))
Subject_Urdu = int(input("Entre your Urdu Subjects Marks Out of 100 : "))

print("Your English mark out of 100 is :",Subject_English,)
print("Your Math mark out of 100 is :",Subject_Math,)
print("Your Science mark out of 100 is :",Subject_Science,)
print("Your Computer mark out of 100 is :",Subject_Computer,)
print("Your Urdu mark out of 100 is :",Subject_Urdu,)
Total_Marks = Subject_English + Subject_Math + Subject_Science + Subject_Computer + Subject_Urdu
print("Your Total Marks out of 500 is :",Total_Marks,)
Percentage = (Total_Marks / 500) * 100
print("Your Percentage is :",Percentage,"%")

```
